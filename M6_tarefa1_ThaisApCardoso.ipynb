#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt

#Carregar o arquivos
arquivo = 'sinasc_RO_2019.csv'
df = pd.read_csv(arquivo)

arquivo_pdf = 'Estrutura_SINASC_para_CD.pdf'
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df = df.astype(str)

#Preparar os dados e criar o gráfico de barras
df[[
    'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE'
]].apply(pd.Series.value_counts).fillna(0).astype(int).plot(kind='bar', stacked=True, colormap='tab20')

plt.title('Distribuição das Variáveis APGAR1, APGAR5, SEXO, GESTACAO e ESTCIVMAE')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.legend(title='Variável')
plt.tight_layout()
plt.show()
```

```python
variables = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

#Criar gráficos de barras para cada variável
for var in variables:
    plt.figure(figsize=(8, 5))
    df[var].value_counts().sort_index().plot(kind='bar', color='skyblue')
    plt.title(f'Distribuição de {var}')
    plt.xlabel(var)
    plt.ylabel('Frequência')
    plt.xticks(rotation=0)

#Ajustando o layout para evitar sobreposição
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
#Calcular a média de APGAR5 por número de CONSULTAS
apgar5_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

#Criar o gráfico de barras
plt.figure(figsize=(10, 6))
apgar5_consultas.plot(kind='bar', color='skyblue')
plt.title('Média de APGAR5 por Número de Consultas Pré-Natais')
plt.xlabel('Número de Consultas Pré-Natais')
plt.ylabel('Média de APGAR5')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
#Configurando o layout do painel com 4 gráficos
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Frequências de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS', fontsize=16)

#Gráfico de frequências para SEXO
df['SEXO'].value_counts().sort_index().plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribuição de SEXO')
axes[0, 0].set_xlabel('SEXO')
axes[0, 0].set_ylabel('Frequência')
axes[0, 0].tick_params(axis='x', rotation=0)

#Gráfico de frequências para GESTACAO
df['GESTACAO'].value_counts().sort_index().plot(kind='bar', ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Distribuição de GESTACAO')
axes[0, 1].set_xlabel('GESTACAO')
axes[0, 1].set_ylabel('Frequência')
axes[0, 1].tick_params(axis='x', rotation=45)

#Gráfico de frequências para ESTCIVMAE
df['ESTCIVMAE'].value_counts().sort_index().plot(kind='bar', ax=axes[1, 0], color='salmon')
axes[1, 0].set_title('Distribuição de ESTCIVMAE')
axes[1, 0].set_xlabel('ESTCIVMAE')
axes[1, 0].set_ylabel('Frequência')
axes[1, 0].tick_params(axis='x', rotation=45)

#Gráfico de frequências para CONSULTAS
df['CONSULTAS'].value_counts().sort_index().plot(kind='bar', ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Distribuição de CONSULTAS')
axes[1, 1].set_xlabel('CONSULTAS')
axes[1, 1].set_ylabel('Frequência')
axes[1, 1].tick_params(axis='x', rotation=0)

#Ajustando o layout para evitar sobreposição
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
